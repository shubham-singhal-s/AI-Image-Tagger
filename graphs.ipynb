```python
auc_acc = []
for l in auc_lines:
    if l =='':
        continue
    x = l.split(' AUC_metric: ')[1]
    x = x.split(' - val_loss:')[0]
    x = float(x)
    auc_acc.append(x)

auc_acc
```

```python
auc_val_acc = []
for l in auc_lines:
    if l =='':
        continue
    x = l.split(' val_AUC_metric: ')[1]
    x = float(x)
    auc_val_acc.append(x)

auc_val_acc
```
